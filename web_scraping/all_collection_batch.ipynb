```python
import pandas as pd
```

```python
# imported the csv files and read it using pd.read_csv

df1 = pd.read_csv('collection_assets_batch_1.csv')
df2 = pd.read_csv('collection_assets_batch_2.csv')
df3 = pd.read_csv('collection_assets_batch_3.csv')
df4 = pd.read_csv('collection_assets_batch_4.csv')
```

```python
df4.head()
```

```python
# Joined the all 4 dataframe(dfs) using concat function

all_batch = pd.concat([df1, df2, df3, df4])
all_batch
```

```python
# Total unique count of all collections

unique_count = len(all_batch["collectionId"].unique())
unique_count
```

```python
# manipulating data stored in a DataFrame called all_batch

collections = all_batch[["collectionId", "collectionName"]]
unique_counts = collections.groupby("collectionId").size().reset_index(name="totalNFTs")

# merge the collections DataFrame with the unique_counts DataFrame using the "collectionId" column as the key
unique_collections = pd.merge(collections, unique_counts, on="collectionId")
unique_value = unique_collections.drop_duplicates() # drops duplicate rows from the unique_collections DataFrame
unique_value
```
